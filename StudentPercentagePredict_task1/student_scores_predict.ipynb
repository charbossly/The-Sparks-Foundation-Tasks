Predict percentage of student based on no of study hours

1- Load data from dataset

```python
import pandas as pd

df = pd.read_csv("student_scores - student_scores.csv")
print(df)
X = df.iloc[0:len(df),0]
y = df.iloc[0:len(df),1]
```

2- Show data 

```python
import matplotlib.pyplot as plt

axes = plt.axes()
axes.grid()
plt.scatter(X,y)
plt.show()
```

3-Split training data and test data

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2, random_state = 0)

print(X_train)
```

4- Regression linear algo

```python
import scipy.stats as stats

slope, intercept,r_value, p_value, str_err = stats.linregress(X_train,y_train);

def predict(x) : 
    return slope * x + intercept


predict_value = predict(X_test)

print(predict_value)
plt.scatter(X,y)
plt.plot(X_test, predict_value, c='g')
plt.show()
```

Answer : Predicted score if student studies for 9.25 is 

```python
print(predict(9.25))
```
